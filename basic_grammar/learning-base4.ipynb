## __name__ == '__main__' python启动入口
__name__ 作为 Python 的魔术内置参数，本质上是模块对象的一个属性

```python
# utils_with_main.py

def get_sum(a, b):
    return a + b
## 作为项目的启动起点，
if __name__ == '__main__':
    print('testing')
    print('{} + {} = {}'.format(1, 2, get_sum(1, 2)))
```
