### Python GIL： python的全局解释器锁
本质上是类似操作系统的 Mutex。每一个 Python 线程，在 CPython 解释器中执行时，都会先锁住自己的线程，阻止别的线程执行。

```python
# 单线程执行
def CountDown(n):
    while n > 0:
        n -= 1
%time CountDown(100000000)        
```

```python
# 多线程执行
from threading import Thread

n = 100000000

t1 = Thread(target=CountDown, args=[n // 2])
t2 = Thread(target=CountDown, args=[n // 2])
t1.start()
t2.start()
t1.join()
t2.join()
```
