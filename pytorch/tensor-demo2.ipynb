```python
# 稀疏张量
i=  torch.tensor([[0,1,1],[2,0,2]]) # 指定了张量值的坐标
v = torch.tensor([1,2,3])
res = torch.sparse_coo_tensor(i,v,(4,4),dtype=torch.float,device=dev)
dres = res.to_dense()
print(res)
print(dres)
```

### 对数运算

```python
# 加法
a = torch.rand(2,3)
b = torch.rand(2,3)
print(a)
print(b)
print(a+b)
print(torch.add(a,b))
print(a.add(b))
# 结果赋值到a
print(a.add_(b))
print(a)
```

```python
# 减法
a = torch.rand(2,3)
b = torch.rand(2,3)
print(a)
print(b)
print(a-b)
print(torch.sub(a,b))
print(a.sub(b))
# 结果赋值到a
print(a.sub_(b))
print(a)
```

```python
# 乘法
a = torch.rand(2,3)
b = torch.rand(2,3)
print(a*b)
print(a.mul(b))
print(torch.mul(a,b))
print(a.mul_(b))
print(a)
```

```python
# 除法
a = torch.rand(2,3)
b = torch.rand(2,3)
print(a/b)
print(torch.div(a,b))
print(a.div(b))
print(a.div_(b))
print(a)
```

```python
# 矩阵运算
a= torch.ones(2,1)
b= torch.ones(1,2)
print(a)
print(b)
print(a @ b)
print(a.matmul(b))
print(a.mm(b))
print(torch.matmul(a,b))
print(torch.mm(a,b))
```

```python
# 高维tensor运算

a = torch.ones(1,2,3,4)
b = torch.ones(1,2,4,3)
print(a)
print(b)
print(a.shape)
print(b.shape)
print(torch.matmul(a,b))
print(torch.matmul(a,b).shape)
```

```python
# 指数运算
a = torch.tensor([1,2])
print(a)
print(a.pow(3))
print(torch.pow(a,3))
print(a**3)
print(a.pow_(3))
print(a)
```

```python
# 自然常数e的运算
a = torch.tensor([1,2],dtype=torch.float)
print(a.type())
print(torch.exp(a))
print(torch.exp_(a))
print(a)
```

```python
# log 运算
a = torch.tensor([1, 2],dtype=torch.float)
print(torch.log(a))
print(torch.log2(a))
print(torch.log10(a))
print(torch.log_(a))
print(a)
```

```python
# 开平方
print(torch.sqrt(a))
print(a.sqrt())
print(torch.sqrt_(a))
print(a)
```

### 广播机制
张量参数可以自动扩展为相同大小
广播机制需要满足的条件：
1. 张量至少有一个维度
2. 满足右对齐,最后一个维度大小相等

```python
import torch

a = torch.rand(2,1,1,3)
b = torch.rand(2,3)
c = a +b
print(a)
print(b)
print(c)
```

### 取余与取整运算

```python
import torch

a = torch.rand(2,2)
a = a * 10
print(a)
#向下取整
print(torch.floor(a))
# 向上取整
print(torch.ceil(a))
# 四舍五入
print(torch.round(a))
# 去除小数部分
print(torch.trunc(a))
# 取小数部分
print(torch.frac(a))
# 对2 取余
print(a % 2)
```
